```python
import requests
import pandas as pd
import re

URL = "https://www.contus.com/careers.php"
EXCEL_NAME = "jobs_manual.xlsx"
COMPANY_NAME = "CONTUS TECH"

response = requests.get(URL, headers={"User-Agent": "Mozilla/5.0"})
response.raise_for_status()

raw_html = response.text
print("RAW HTML (first 1000 chars):\n")
print(raw_html[:1000])

# Split raw HTML by newlines and spaces
lines = raw_html.split("\n")
lines = [line.strip() for line in lines if line.strip()]

# Only lines that might contain jobs (filtering keywords)
job_lines = [line for line in lines if "Apply Now" in line or "Senior" in line or "Engineer" in line or "Developer" in line or "Analyst" in line]

print("Potential job lines (first 20):")
print(job_lines[:20])

jobs = []

# The page structure: Job title on one line, Location next, "Apply Now" next
for i in range(len(job_lines)):
    if "Apply Now" in job_lines[i]:
        # Job title = 2 lines above
        if i >= 2:
            job_title = job_lines[i-2]
            location = job_lines[i-1]
            description = f"{job_title} role at {COMPANY_NAME} located in {location}"
            jobs.append({
                "job_title": job_title,
                "company_name": COMPANY_NAME,
                "location": location,
                "job_description": description
            })

print(f"Extracted {len(jobs)} jobs (first 5):")
jobs[:5]
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

URL = "https://www.contus.com/careers.php"
EXCEL_NAME = "jobs_fully_manual.xlsx"
COMPANY_NAME = "CONTUS TECH"

# Step 1: Fetch HTML
response = requests.get(URL, headers={"User-Agent": "Mozilla/5.0"})
response.raise_for_status()
html = response.text

# Step 2: Parse HTML
soup = BeautifulSoup(html, "html.parser")

# Step 3: Find all job blocks
# On Contus page, jobs are listed under a <div> that contains h4 (title) + location text + "Apply Now"
# We'll loop through h4 tags and look for the next text and next <a> tag
jobs = []

for h4 in soup.find_all("h4"):
    job_title = h4.get_text(strip=True)

    # Get location: next sibling that is text (ignore empty lines)
    location_tag = h4.find_next_sibling(text=True)
    while location_tag and not location_tag.strip():
        location_tag = location_tag.find_next_sibling(text=True)
    location = location_tag.strip() if location_tag else "Not specified"

    # Get "Apply Now" link
    apply_tag = h4.find_next("a", string=lambda s: "Apply" in s)
    apply_link = apply_tag['href'] if apply_tag else "Not specified"

    # Manual description
    description = f"{job_title} role at {COMPANY_NAME} located in {location}"

    # Only add if apply link exists (filter out headings like "Join Us")
    if apply_link != "Not specified":
        jobs.append({
            "job_title": job_title,
            "company_name": COMPANY_NAME,
            "location": location,
            "job_description": description,
            "apply_link": apply_link
        })

# Step 4: Show first 5 jobs
print(f"Extracted {len(jobs)} jobs (first 5):")
jobs[:5]

```
