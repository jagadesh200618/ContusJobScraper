```python
# Class definition
class Student:
    def __init__(self, name, roll_no):
        self.name = name
        self.roll_no = roll_no

    def display(self):
        print("Name:", self.name)
        print("Roll No:", self.roll_no)

# Object creation
s1 = Student("Jagadesh", 101)

# Method call
s1.display()
```
